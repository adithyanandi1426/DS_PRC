```
#ingest circuits.csv

from pyspark.sql.types import StructType, IntegerType, StructField, StringType, DoubleType, NullType
```

```
health_schema = StructType(fields=[StructField("Entity", StringType(), False ),
                                    StructField("Year", IntegerType(), True ),
                                    StructField("Share of cancer deaths attributable to risk factor", DoubleType(), True)
                                   ])
```

```
health_df = spark.read \
.option("header", True) \
.schema(health_schema) \
.csv("/FileStore/tables/share_of_cancer_deaths_attributed_to_risk_factors.csv")
```

```
from pyspark.sql.functions import col
```

```
health_select = health_df.select(col("Entity"),
                                col("Year"),
                                col("Share of cancer deaths attributable to risk factor"))
```

```
health_rename = health_select.withColumnRenamed("Entity", "entity")\
.withColumnRenamed("Year", "year")\
.withColumnRenamed("Share of cancer deaths attributable to risk factor", "risk factor")
```

```
# write data to datalae as parquet

health_rename.write.mode("overwrite").parquet("/mnt/formula1idl/processed/health")
```
